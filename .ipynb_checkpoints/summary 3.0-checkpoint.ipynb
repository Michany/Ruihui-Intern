```python
p=s.signal_class_list[0].profit
v=s.signal_class_list[0].volume
```

```python
import pandas as pd
a=pd.DataFrame({'Profit': list(p.values())}, index=pd.MultiIndex.from_product(
            [[s.signal_class_list[0].name], p.keys()]))
b=pd.DataFrame({'Volume': list(v.values())}, index=pd.MultiIndex.from_product(
            [[s.signal_class_list[0].name], v.keys()]))
joint=a.join(b)
```

```python
joint['P/V']=joint['Profit']/joint['Volume']
```

```python
for index in joint.index:
    new.append(list(joint.xs(index[0],level=0).xs(index[1])))
```
